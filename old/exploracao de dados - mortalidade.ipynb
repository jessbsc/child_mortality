```python
data = pd.read_csv("mortality_data.csv")
```

```python
data= data.drop(['Unnamed: 0'], axis=1)
```

```python
data_nan= data[pd.isna(data.LOCAL_NAME)]
```

```python
local_code_to_remove = data_nan['LOCAL_CODE'].unique()
```

```python
data_clean = data[~data['LOCAL_CODE'].isin(local_code_to_remove)]
```

```python
fig = plt.figure(figsize =(10, 7))
plt.boxplot(data_clean['POPULATION'])
 
# show plot
plt.show()
```

```python
data_2021= data_clean[data_clean.YEAR==2021]
```

```python
data_2021
```

```python
data_2021.describe()
```

```python
data_2021.sort_values(by=['POPULATION_BIRTH'], ascending=False)[['LOCAL_NAME','POPULATION_BIRTH']]
```

```python
data_clean[['POPULATION','POPULATION_BIRTH']].describe([.01,.1,.2,.3,.4,.5,.6,.7,.8,.9,.99])
```

```python
data_filter = data_clean[data_clean.POPULATION_BIRTH>60]
```

```python
data_filter = data_filter[data_filter.POPULATION_BIRTH<871]
```

```python
data_filter.describe()
```

```python
fig = plt.figure(figsize =(10, 7))
plt.boxplot(data_filter['POPULATION'])
 
# show plot
plt.show()
```

```python
plt.boxplot(data_filter['POPULATION_BIRTH'])
 
# show plot
plt.show()
```

```python
data_agg = data_filter[['YEAR', 'POPULATION_BIRTH', 'POPULATION']].groupby('YEAR').agg('sum')
data_agg
```

```python
data_agg = data_agg.reset_index()
```

```python
data_agg_without_filter = data_clean[['YEAR', 'POPULATION_BIRTH', 'POPULATION']].groupby('YEAR').agg('sum')
data_agg_without_filter = data_agg_without_filter.reset_index()
data_agg_without_filter
```

```python
data_format = data_format[data_format.LOCAL_CODE!=110000]
data_format = data_format[~data_format.year_month_date.isnull()]
```

```python
data_format = data_format[data_format.year_month_date<'2022-01-01']
```

```python
lista_data =  data_format['year_month_date'].drop_duplicates().sort_values()
lista_data_clean = list(lista_data)
```

```python
lista_data_clean
```

```python
lista_local_code =  list(data_format['LOCAL_CODE'].drop_duplicates())
```

```python
for code in lista_local_code: 
    data_filter = data_format[data_format['LOCAL_CODE']==code]
    data_filter = data_filter.fillna("")
    data_filter['grupo'] = data_filter['grupo'].astype(str)
    lista_data_city =  list(data_filter['year_month_date'].drop_duplicates().sort_values())
    diference = list(set(lista_data_clean)-set(lista_data_city))
    if len(diference) > 0: 
        df = pd.DataFrame(diference, columns =['year_month_date'])
        data_filter = data_filter.fillna("")
        #df['LOCAL_NAME'] = max(data_filter.LOCAL_NAME)
        #df['grupo'] = max(data_filter.grupo)
        #df['LOCAL_CODE'] = max(data_filter.LOCAL_CODE)
        data_format= data_format.append(df)

   
```

```python
data_format['LOCAL_NAME'] = data_format['LOCAL_NAME'].fillna("")
data_format['grupo'] = data_format['grupo'].fillna(0)
for code in lista_local_code: 
    data_format.loc[data_format['LOCAL_CODE'] == code, 'LOCAL_NAME'] = max(data_format[data_format['LOCAL_CODE'] == code]['LOCAL_NAME'])
    data_format.loc[data_format['LOCAL_CODE'] == code, 'grupo'] = max(data_format[data_format['LOCAL_CODE'] == code]['grupo'])
```

```python
lista_uf =  list(data_format['sigla_uf'].drop_duplicates())

for uf in lista_uf: 
    data_format.loc[data_format['sigla_uf'] == uf, 'area_total'] = mean(data_format[data_format['sigla_uf'] == uf]['area_total'])
```
